# Visualization of continuous data using Histogram

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Advertising.csv")
data
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['sales'], bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of Sales')
plt.xlabel('Sales')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
